```python
import nltk
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords
nltk.download('punkt')
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Climate_change').read() #getting the data source

#parsing the data
soup = bs.BeautifulSoup(source,'lxml')
```

```python
#Fetching the data
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
#preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',text)
```

```python
text = re.sub(r'\s+',' ',text)
```

```python
text = text.lower()
```

```python
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s',' ',text)
```

```python
sentences = nltk.sent_tokenize(text)

sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```
